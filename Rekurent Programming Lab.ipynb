```python
#Задача 16
a = 'ptslddf'
b = 'tsgldds'

print('a =', a)
print('b =', b)
print('len a =', len(a))
print('len b =', len(b))


def distance(a, b):
    "Calculates the Levenshtein distance between a and b."
    n, m = len(a), len(b)
    if n > m:
        # Make sure n <= m, to use O(min(n, m)) space
        a, b = b, a
        n, m = m, n

    current_row = range(n + 1)  # Keep current and previous row, not entire matrix
    for i in range(1, m + 1):
        previous_row, current_row = current_row, [i] + [0] * n
        for j in range(1, n + 1):
            add, delete, change = previous_row[j] + 1, current_row[j - 1] + 1, previous_row[j - 1]
            if a[j - 1] != b[i - 1]:
                change += 1

            current_row[j] = min(add, delete, change)

    return current_row[n]

print('Количество действий равно', distance(a, b))
```

```python
def seq_delmaxs(array, num):
    ds = [0 for i in range(num)]
    ps = [0 for i in range(num)]
    
    for i in range(num):
        ds[i] = 1
        ps[i] = -1
        for j in range(i):
            if array[j] < array[i]:
                if ds[j]+1 > ds[i]:
                    ds[i] = ds[j] + 1
                    ps[i] = j
                    
    pos = 0
    for i in range(1, num):
        if ds[i] > ds[pos]:
            pos = i

    j = 0

    while pos != -1:
        ds[j] = array[pos]
        j += 1
        pos = ps[pos]

    for i in range(j):
        array[i] = ds[j-1-i]

    return j

def main_function():
    array = [70, 1, 2, -4, 3, 10, 6, 5, 6, 0, 9, 1]
    array_len = len(array)
    n = seq_delmaxs(array, array_len)

    for i in range(n):
        print(array[i], end=' ')
        
main_function()
```
